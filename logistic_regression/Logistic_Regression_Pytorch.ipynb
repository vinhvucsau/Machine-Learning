# Logistic Regression with Pytorch

```python
import numpy as np
import torch 
import torch.nn as nn 

# Data preparation
data = np.genfromtxt('../datasets/iris_1D.csv',delimiter=',',skip_header=1)
x_data = torch.from_numpy(data[:, 0:1]).float()
y_data = torch.from_numpy(data[:,1:]).float()

# Create model , Loss and Optimizer
linear = nn.Linear(1,1)
# Binary Cross Entropy
loss_fn = nn.BCELoss()
optimizer = torch.optim.SGD(linear.parameters(),lr= 0.01)
# Training
epochs =100 
for epoch in range(epochs):
    
    #Sigmoid Function
    y_hat = torch.sigmoid(linear(x_data))
    
    loss =loss_fn(y_hat,y_data)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

### Logistic Regression using logit no compute y_hat  

```python
import numpy as np
import torch 
import torch.nn as nn 

# Data preparation
data = np.genfromtxt('../datasets/iris_2D_demo.csv',delimiter=',',skip_header=1)
x_data = torch.from_numpy(data[:, 0:2]).float()
y_data = torch.from_numpy(data[:,2:]).float()

# Create model , Loss and Optimizer
# replace 1 to 2 
linear = nn.Linear(2,1)
# Binary Cross Entropy
loss_fn = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(linear.parameters(),lr= 0.01)
losses = []
# Training
epochs =100 
for epoch in range(epochs):
    
    #Sigmoid Function
    z= linear(x_data)
    
    loss =loss_fn(z,y_data)
    losses.append(loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
import matplotlib.pyplot as plt 
plt.plot(losses)
```
